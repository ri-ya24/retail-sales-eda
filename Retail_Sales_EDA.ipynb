## Region-wise Sales Analysis

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))

region_sales.plot(kind="bar", color="lightgreen")

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)

plt.show()
```

### Business Insight

## Top 10 States by Sales

```python
state_sales = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

state_sales
```

```python
plt.figure(figsize=(10,6))

state_sales.plot(kind="bar", color="orange")

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.show()
```

### Business Insight

- California generated the highest total sales, making it the company's strongest market.
- Texas ranked second, followed by New York.
- The top 10 states contributed a significant share of overall sales.
- These states can be prioritized for marketing campaigns, inventory planning, and business expansion.

## Top 10 Products by Sales

```python
product_sales = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

product_sales
```

```python
plt.figure(figsize=(12,6))

product_sales.plot(kind="bar", color="purple")

plt.title("Top 10 Products by Sales")
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)

plt.show()
```

### Business Insight

- Riverside Palais Royal Lawyers Bookcase generated the highest total sales.
- Atlantic Metals Mobile 5-Shelf Bookcases ranked second.
- Canon imageCLASS 2200 Advanced Copier was among the top-performing technology products.
- The top 10 products contributed a significant portion of overall sales, indicating that a few high-performing products drive a large share of revenue.

## Monthly Sales Trend

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)

monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()

monthly_sales
```

```python
plt.figure(figsize=(12,6))

monthly_sales.plot(kind="line", marker="o", color="green")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)

plt.show()
```

### Business Insight

- Sales fluctuated across different months, indicating changes in customer demand over time.
- Several months showed significantly higher sales, suggesting seasonal peaks.
- Analyzing monthly trends helps businesses plan inventory, marketing campaigns, and future sales forecasts.

## Sales by Customer Segment


```python
segment_sales = df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)

segment_sales
```

```python
plt.figure(figsize=(8,5))

segment_sales.plot(kind="bar", color="teal")

plt.title("Total Sales by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)

plt.show()
```

### Business Insight

- The Consumer segment generated the highest total sales.
- The Corporate segment ranked second in terms of sales.
- The Home Office segment contributed the lowest sales.
- The company should continue focusing on the Consumer segment while exploring growth opportunities in the Home Office segment.

## Sales Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=30)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

### Business Insight

- Most sales transactions are concentrated in the lower sales range.
- A few transactions have exceptionally high sales values, indicating possible outliers.
- Understanding the sales distribution helps identify customer purchasing patterns.
